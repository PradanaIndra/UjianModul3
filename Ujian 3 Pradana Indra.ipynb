### Handling Missing Values
The first one to handle is the country missing value. as we can see, there is only 1 missing value that needed to be handled. Based on the code below, the MOST country appeared is PRT or we can say portugal in the number of 1774. so to handling the missing value we using the MODUS of the Data to fill in.

```python
hb['country'].value_counts()
```

```python
hb['country'] = hb['country'].fillna('PRT')
```

```python
hb.isna().sum()
```

```python
hb['agent'].value_counts()
```

### Handling missing values on Agent

According to the feature, Agent is a categorical type of data. The description is Agent is an ID of the travel agency that made the booking. So if i fill the missing values with the median of it, it will ruin the Travel Agency that did not make any booking. So the best thing i can do is to let the missing values keep missing or replace it with 0 instead.

```python
hb['agent'] = hb['agent'].fillna(0)
```

```python
hb.isna().sum()
```

### Handling Company missing values

Due to the 96% Company is a missing value data, to avoid too much outliers if i put the new spesifik ID such a null (0) tu replace the nan. i decided to drop the entire column to make the data better

```python
hb['company'].value_counts()
```

```python
hb = hb.drop(['company'], axis=1)
```

```python
hb.isna().sum()
```

## Boxplot & Barplot

**required car parking spaces**

as we see from the boxplot below, required car parking spaces is the Number of car parking spaces required by the customer. If the place have less parking lot or a small parking lot, they are mostlikely to cancel that.

```python
plt.figure(figsize = (8,8))
sns.boxplot(x = 'required_car_parking_spaces', y = 'is_canceled', data = hb)
```

```python
plt.figure(figsize = (8,8))
sns.barplot(x = 'required_car_parking_spaces', y = 'is_canceled', data = hb)
```

**Lead Time**

Lead time is the number of days that elapsed between the entering date of the booking and the arrival date. So as we can see, the number of people canceling the booking is mostlikely high on both range date. between 0 and 50. But people tend to cancel the booking if they book more than 50 days before the arriving date.

```python
plt.figure(figsize = (8,8))
sns.barplot(x = 'is_canceled', y = 'lead_time', data = hb)
```

**Booking changes**

Booking changes is the number of changes/amendments made to the booking from the moment the booking was entered until the moment of check-in or cancellation. in story short, booking changes is reschedule. People most likely tend to not cancel the booking if they make a booking changes/reschedule.

```python
plt.figure(figsize = (8,8))
sns.boxplot(x = 'booking_changes', y = 'is_canceled', data = hb)
```
